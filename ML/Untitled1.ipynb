```python
import translators as ts
ts._google.language_map

def text_to_sentences(text):
    result = re.sub(r'(\S)\.(\S)', r'\1$\2', text) # mengganti (.) yang bukan sebagai pemisah kalimat dengan ($)
    result = re.findall(r'\S[^\.]*\.', result) # memisahkan paragraf menjadi kalimat
    result = [re.sub(r'(\S)\$(\S)', r'\1.\2', r) for r in result] # mengganti ($) kembali ke (.)
    dataIndo = []
    for i in result:
        try:
            dataIndo.append(ts.google(i, from_language='en', to_language='id'))
        except:
            dataIndo.append('')
    print('done')
    return dataIndo

def text_to_sentences2(text):
    result = re.sub(r'(\S)\.(\S)', r'\1$\2', text) # mengganti (.) yang bukan sebagai pemisah kalimat dengan ($)
    result = re.findall(r'\S[^\.]*\.', result) # memisahkan paragraf menjadi kalimat
    result = [re.sub(r'(\S)\$(\S)', r'\1.\2', r) for r in result] # mengganti ($) kembali ke (.)
    return result
```

```python
stopWords = []
factory = StemmerFactory()
stemmer = factory.create_stemmer()

#start replaceTwoOrMore
def replaceTwoOrMore(s):
    #look for 2 or more repetitions of character and replace with the character itself
    pattern = re.compile(r"(.)\1{1,}", re.DOTALL)
    return pattern.sub(r"\1\1", s)
#end

#start getStopWordList
def getStopWordList():
    #read the stopwords file and build a list
    stopWords = StopWordRemoverFactory()
    more_stopword = ['dengan', 'ia','bahwa','oleh','AT_USER','URL','di','yg','dari','ke','ini','bgmn','tmn2','dr','pt','dg']
    data = stopWords.get_stop_words()+more_stopword
    return data
#end

def steaming_text(sentence):
    return stemmer.stem(sentence)

def getFeatureVector(tweet):
    featureVector = []
    #split tweet into words
    words = tweet.split()
    for w in words:
        #replace two or more with two occurrences
        w = replaceTwoOrMore(w)
        #strip punctuation
        w = w.strip('\'"?,.')
        #check if the word stats with an alphabet
        val = re.search(r"^[a-zA-Z][a-zA-Z0-9]*$", w)
        #ignore if it is a stop word
        if(w in stopWords or val is None):
            continue
        else:
            featureVector.append(w.lower())
    return featureVector

def getFeatureVectorJournal(journal):
    journalset = text_to_sentences2(str(journal))
    sentence_list = []
    for i in journalset:
        vektor = getFeatureVector(i)
        sentence_list.append(steaming_text(listToString(vektor))+'.')
    return sentence_list

stopWords = getStopWordList()
```

```python
def normalization(embeds):
  norms = np.linalg.norm(embeds, 2, axis=1, keepdims=True)
  return embeds/norms

# cosine similarity (all sentence at once)
def similarity(a, b):
  return np.matmul(a, np.transpose(b))
```
